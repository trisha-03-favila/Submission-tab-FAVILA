```python
# Data Structures Activity Part 2
# Student: Trisha Ann Favila
# Student Number: 202340700

# 1. Creating Matrices

# 1.a.i. 1st Matrix based on initials
first_set = [20, 1, 6]  # T, A, F
second_set = [1, 2, 3]  # Example second letters
first_matrix = [first_set, second_set]

# 1.a.ii. 2nd Matrix based on student number
first_numbers = [2, 0, 2]  # from 202340700
last_numbers = [7, 0, 0]
second_matrix = [first_numbers, last_numbers]

# 1.b. Print both matrices
print("First Matrix:")
print(first_matrix)
print("\nSecond Matrix:")
print(second_matrix)

# 2. Matrix Addition
third_matrix = [
    [first_matrix[0][i] + second_matrix[0][i] for i in range(3)],
    [first_matrix[1][i] + second_matrix[1][i] for i in range(3)]
]

print("\nThird Matrix (Addition Result):")
print(third_matrix)

# 3. Scalar Multiplication
scalar = 2
fourth_matrix = [
    [element * scalar for element in row] for row in first_matrix
]

print("\nFourth Matrix (Scalar Multiplication of First Matrix by 2):")
print(fourth_matrix)

# 4. Matrix Transpose
fifth_matrix = [
    [second_matrix[0][0], second_matrix[1][0]],
    [second_matrix[0][1], second_matrix[1][1]],
    [second_matrix[0][2], second_matrix[1][2]]
]

print("\nFifth Matrix (Transpose of Second Matrix):")
print(fifth_matrix)

# 5. Matrix Multiplication (Third Matrix x Fifth Matrix)
sixth_matrix = [
    [
        third_matrix[0][0]*fifth_matrix[0][0] + third_matrix[0][1]*fifth_matrix[1][0] + third_matrix[0][2]*fifth_matrix[2][0],
        third_matrix[0][0]*fifth_matrix[0][1] + third_matrix[0][1]*fifth_matrix[1][1] + third_matrix[0][2]*fifth_matrix[2][1]
    ],
    [
        third_matrix[1][0]*fifth_matrix[0][0] + third_matrix[1][1]*fifth_matrix[1][0] + third_matrix[1][2]*fifth_matrix[2][0],
        third_matrix[1][0]*fifth_matrix[0][1] + third_matrix[1][1]*fifth_matrix[1][1] + third_matrix[1][2]*fifth_matrix[2][1]
    ]
]

print("\nSixth Matrix (Matrix Multiplication Result):")
print(sixth_matrix)

# 6. Sum of All Elements in Third Matrix
sum_third_matrix = sum(sum(row) for row in third_matrix)

print("\nSum of all elements in Third Matrix:")
print(sum_third_matrix)

# 7. Zero Matrix
seventh_matrix = [
    [0, 0, 0],
    [0, 0, 0]
]

print("\nSeventh Matrix (Zero Matrix 2x3):")
print(seventh_matrix)

```
